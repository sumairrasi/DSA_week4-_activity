# 1. Compare churn count with respect to gender.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
sns.countplot(x= "gender", hue = "Churn", data = data)
plt.title('Churn count on male',fontsize=15)
plt.ylabel('Churn Count')
plt.xticks()
plt.yticks()
```

# 2. Find out how many female senior citizens there in the dataset

```python
data3 = data[data['SeniorCitizen']==1.0]
```

```python
data4 = data3[data3['gender']=='Female']
```

```python
data5 = data4.gender
```

```python
data9 = data5.count()
```

```python
data8 = data[data['gender'] == 'Female']
```

```python
sns.countplot(x= "gender", hue = "Churn", data = data8)
plt.title('Churn count on male',fontsize=15)
plt.ylabel('Churn Count')
plt.xticks()
plt.yticks()
```

```python
print('Female citizens are:',data9)
```
